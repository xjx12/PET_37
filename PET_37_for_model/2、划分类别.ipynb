```python
import os
import shutil
```

```python
# 定义函数来分类并移动图片  
def classify_and_move_images(input_folder):    
    for filename in os.listdir(input_folder):  # 遍历文件夹中的所有文件
        if filename.endswith('.jpg'):   # 检查文件是否为jpg图片
            file_path = os.path.join(input_folder, filename)  # 获取图片完整路径
            # 因为有些宠物名字中有下划线，所以用最后一个下划线作为分割依据
            last_underscore_index = filename.rfind('_')  #找到图片名中的最后一个下划线位置
            if last_underscore_index != -1:  # 确保找到了一个下划线  
                category = filename[:last_underscore_index]  # 分割图片名和类别 
                category_folder = os.path.join(input_folder, category)  # 创建或确保分类文件夹存在 
                if not os.path.exists(category_folder):  
                    os.makedirs(category_folder)  
                    
                new_file_path = os.path.join(category_folder, filename)   # 将图片移动到对应的分类文件夹中且保持文件名不变
                shutil.move(file_path, new_file_path)  
                print(f'Moved {file_path} to {new_file_path}')  
            else:  
                print(f'No underscore found in {filename}, skipping...')  
```

```python
# 设置两个图片所在的文件夹路径  
train = './oxford-iiit-pet/train'  
test = './oxford-iiit-pet/test' 
# 调用函数处理1、中处理好的两个文件夹  
classify_and_move_images(train)  
classify_and_move_images(test)
```

```python
def splitImg2Category(dataDir="./oxford-iiit-pet/images/", resDir="dataset/"):
#归类图像到不同目录中
    for one_pic in os.listdir(dataDir):
        one_path = os.path.join(dataDir, one_pic)
        class_name = one_pic.split('_')[0].strip()
        oneDir = os.path.join(resDir, class_name)
        if not os.path.exists(oneDir):
            os.makedirs(oneDir)
        shutil.copy(one_path, os.path.join(oneDir, one_pic))
```
